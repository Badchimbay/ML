<img src="https://i.ibb.co/93gy1dN/Screenshot-9.png" width="200">

Создать полносвязный слой с 3 нейронами и прогнать через него батч `inputs`. Полученный результат пропустить через функцию активации softmax. Посчитать значение CCE, трактуя вектор `y` как вектор правильных ответов.

```python
class CategoricalCrossentropyLoss:
  def forward(self, y_pred, y_true):
    # <реализовать логику CCE>
    eps = torch.finfo(float).eps
    y_pred = torch.clip(y_pred, eps, 1. - eps)
    return -torch.sum(y_true * torch.log(y_pred + eps)) / y_pred.shape[0]
```

2.3.3 Модифицировать 2.3.1, добавив L2-регуляризацию.

![](https://wikimedia.org/api/rest_v1/media/math/render/svg/d92ca2429275bfdc0474523babbafe014ca8b580)


```python
class MSELossL2:
  def __init__(self, lambda_):
    # <создать атрибут объекта lambda_>
    self.lambda_ = lambda_

  def data_loss(self, y_pred, y_true):
    # <подсчет первого слагаемого из формулы>
    return torch.mean(torch.pow(y_true - y_pred, 2))

  def reg_loss(self, layer):
    # используйте атрибуты объекта layer, в которых хранятся веса слоя
    # <подсчет второго слагаемого из формулы>
    return self.lambda_ * torch.sum(torch.pow(layer.weights, 2))

  def forward(self, y_pred, y_true, layer):
    return self.data_loss(y_pred, y_true) + self.reg_loss(layer)
```

```python
inputs = torch.tensor([[1, 2, 3, 2.5], 
                       [2, 5, -1, 2], 
                       [-1.5, 2.7, 3.3, -0.8]])

y = torch.tensor([2, 3, 4])
```

```python
class SquaredLoss:
  def forward(self, y_pred, y_true):
    return torch.mean(torch.pow(y_true - y_pred, 2)) # <реализовать логику MSE>

  def backward(self, y_pred, y_true):
    self.dinput = torch.mean(-2.0*(y_true - y_pred)).repeat(y_pred.shape[0])
```

2.4.1.2. Модифицируйте класс `Neuron` из __2.1.1__:

```python
class Neuron:
  def __init__(self, n_inputs):
    # <создать атрибуты объекта weights и bias>
    self.weights = torch.nn.parameter.Parameter(torch.normal(0, 1, (n_inputs, )))
    self.bias = torch.nn.parameter.Parameter(torch.normal(0, 1, (n_inputs, )))
  
  def forward(self, inputs):
    self.inputs = inputs
    return torch.matmul(inputs, self.weights) + self.bias # <реализовать логику нейрона>
  
  def backward(self, dvalue):
    # dvalue - значение производной, которое приходит нейрону от следующего слоя сети
    # в данном случае это будет значение df/dc (созданное методом backwards у объекта MSELoss)
    self.dweights = torch.matmul(self.inputs.T, dvalue) # df/dW
    self.dbias = torch.sum(dvalue, axis=0, keepdims=True) # df/dB
    self.dinput = torch.dot(dvalue, self.weights.T) # df/wX
```

```python
class Linear:
  def __init__(self, n_features, n_neurons):
    # <создать атрибуты объекта weights и biases>
    self.n_features = n_features
    self.n_neurons = n_neurons
    self.weights = torch.nn.parameter.Parameter(torch.normal(0, 1, (n_features, n_neurons)))
    self.bias = torch.nn.parameter.Parameter(torch.normal(0, 1, (n_neurons, )))
  
  def forward(self, inputs):
    self.inputs = inputs
    return torch.matmul(inputs, self.weights) + self.bias # <реализовать логику слоя>

  def backward(self, dvalue):
    self.dweights = torch.matmul(self.inputs.T, dvalue) # df/dW
    self.dbias = torch.sum(dvalue, axis=0, keepdims=True) # df/dB
    self.dinput = torch.matmul(dvalue, self.weights.T) # df/wX
```
